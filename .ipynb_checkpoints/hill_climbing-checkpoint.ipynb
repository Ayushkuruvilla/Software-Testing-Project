```python
import numpy as np
import pandas as pd
import onnxruntime as rt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from skl2onnx.common.data_types import FloatTensorType
```

```python
data = pd.read_csv('data/investigation_train_large_checked.csv')
data = data.drop(columns=['Ja', 'Nee'])
y = data['checked']
X = data.drop(['checked'], axis=1)
X = X.astype(np.float32)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True, stratify=y)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
unique, counts = np.unique(y_test, return_counts=True)
print("Binary value distribution in y_test:", dict(zip(unique, counts)))
```

```python
import numpy as np
import onnxruntime as rt
from concurrent.futures import ThreadPoolExecutor, as_completed
from multiprocessing import Manager
def get_predict_label_and_confidence(session, solution):
    """Runs inference and returns the predicted label with confidence scores."""
    y_pred_solution = session.run(None, {'X': solution.astype(np.float32)})

    label = y_pred_solution[0][0]  # Predicted class
    confidence_top1_label = y_pred_solution[1][0][0]
    confidence_top2_label = y_pred_solution[1][0][1]

    return label, confidence_top1_label, confidence_top2_label

def get_mean_and_variance(feature):
    """Returns the mean and variance of a dataset feature."""
    return feature.mean(), feature.var()

def get_fitness(y_true, y_pred):
    """Calculates fitness based on prediction confidence."""
    label_mutated, confidence_top1, confidence_top2 = y_pred
    return -confidence_top1 if label_mutated != y_true else confidence_top1 - confidence_top2
```

```python


def random_gaussian_mutation(solution, mutation_rate, feature_stats):
    """Applies Gaussian mutations to a subset of features in the solution."""
    num_features = solution.shape[1]
    mutation_rate = np.random.uniform(0.1, 0.5)  # Random mutation rate between 10%-50%
    num_mutations = max(1, int(mutation_rate * num_features))

    mutated_solution = solution.copy()

    # Select random features to mutate
    feature_indices = np.random.choice(num_features, num_mutations, replace=False)

    # Apply mutations using precomputed mean/variance
    for idx in feature_indices:
        mean, variance = feature_stats[idx]
        mutation_value = np.random.normal(loc=mean, scale=np.sqrt(variance))
        mutated_solution[0, idx] += mutation_value if np.random.rand() < 0.5 else -mutation_value

    return mutated_solution

def stochastic_hill_climber(initial_solution, model, y_test_sample, num_iterations=10000, num_neighbors=10):
    """Performs stochastic hill climbing to optimize the solution."""
    session = rt.InferenceSession(f"Group19/model_{model}.onnx")
    current_solution = initial_solution
    current_prediction = get_predict_label_and_confidence(session, current_solution)
    current_fitness = get_fitness(y_test_sample, current_prediction)

    feature_stats = [(get_mean_and_variance(X_test.iloc[:, i])) for i in range(X_test.shape[1])]

    counter = 0
    for i in range(num_iterations):
        # Generate mutated solutions
        neighbors = [random_gaussian_mutation(current_solution, 0.1, feature_stats) for _ in range(num_neighbors)]
        
        # Evaluate fitness in bulk to reduce inference calls
        predictions = [get_predict_label_and_confidence(session, neighbor) for neighbor in neighbors]
        neighbor_scores = [get_fitness(y_test_sample, pred) for pred in predictions]

        # Find the best mutated neighbor
        best_neighbor_index = np.argmin(neighbor_scores)
        best_neighbor_score = neighbor_scores[best_neighbor_index]

        # Accept the best neighbor if it improves the solution
        if best_neighbor_score < current_fitness:
            current_solution = neighbors[best_neighbor_index]
            current_fitness = best_neighbor_score

            # Track prediction flips
            new_prediction = predictions[best_neighbor_index]
            if current_prediction[0] != new_prediction[0]:
                nFlips[model] += 1
                if counter == 0:
                    flippedIterations[model].append(i+1)
                    counter += 1

    return current_solution, current_fitness

# Shared data structures for tracking flips
manager = Manager()
nFlips = manager.dict({1: 0, 2: 0})
flippedIterations = manager.dict({1: [], 2: []})

best_solutions_scores_1 = []
best_solutions_scores_2 = []

# Select a subset of test data
X_test_subset = X_test.values  # Convert DataFrame to NumPy for faster processing
y_test_subset = y_test.values

def task(X_test_sample, y_test_sample, model):
    """Runs stochastic hill climbing for a single sample."""
    initial_solution = X_test_sample.reshape(1, -1)
    final_solution, best_score = stochastic_hill_climber(initial_solution, model, y_test_sample, num_iterations=10, num_neighbors=10)
    return X_test_sample, y_test_sample, final_solution, best_score

# Parallel Execution Using ProcessPoolExecutor
tasks = [(X_test_sample, y_test_sample, 1) for X_test_sample, y_test_sample in zip(X_test_subset, y_test_subset)]

with ThreadPoolExecutor() as executor:
    running_tasks = [executor.submit(task, *task_args) for task_args in tasks]
    for running_task in as_completed(running_tasks):
        best_solutions_scores_1.append(running_task.result())

# Repeat for Model 2
tasks = [(X_test_sample, y_test_sample, 2) for X_test_sample, y_test_sample in zip(X_test_subset, y_test_subset)]

with ThreadPoolExecutor() as executor:
    running_tasks = [executor.submit(task, *task_args) for task_args in tasks]
    for running_task in as_completed(running_tasks):
        best_solutions_scores_2.append(running_task.result())
```

```python
# Columns to mutate
TARGET_FEATURES = ["persoon_leeftijd_bij_onderzoek", "persoon_geslacht_vrouw"]

def random_gaussian_mutation_specific(solution, mutation_rate, feature_stats, feature_indices):
    """Applies Gaussian mutation only to specific columns (age and gender)."""
    mutated_solution = solution.copy()

    # Apply mutation only to the specified columns
    for feat, idx in feature_indices.items():
        mean, variance = feature_stats[feat]

        if feat == "persoon_leeftijd_bij_onderzoek":
            # Apply Gaussian noise to age
            mutation_value = np.random.normal(loc=mean, scale=np.sqrt(variance))
            mutated_solution[0, idx] = max(0, min(100, mutated_solution[0, idx] + mutation_value))

        elif feat == "persoon_geslacht_vrouw":
            # Boolean flip for gender
            mutated_solution[0, idx] = 1 - mutated_solution[0, idx]  # Flip 0 <-> 1

    return mutated_solution

def stochastic_hill_climber_specific(initial_solution, model, y_test_sample, num_iterations=100, num_neighbors=10):
    """Performs stochastic hill climbing only on age and gender features."""
    session = rt.InferenceSession(f"Group19/model_{model}.onnx")
    current_solution = initial_solution
    current_prediction = get_predict_label_and_confidence(session, current_solution)
    current_fitness = get_fitness(y_test_sample, current_prediction)

    # Compute mean & variance only for targeted features
    feature_indices = {feat: X_test.columns.get_loc(feat) for feat in TARGET_FEATURES}
    feature_stats = {feat: get_mean_and_variance(X_test.iloc[:, idx]) for feat, idx in feature_indices.items()}

    counter = 0
    for i in range(num_iterations):
        # Generate mutated solutions (only modifying selected features)
        neighbors = [random_gaussian_mutation_specific(current_solution, 0.1, feature_stats, feature_indices) for _ in range(num_neighbors)]
        
        # Evaluate fitness in bulk to reduce inference calls
        predictions = [get_predict_label_and_confidence(session, neighbor) for neighbor in neighbors]
        neighbor_scores = [get_fitness(y_test_sample, pred) for pred in predictions]

        # Find the best mutated neighbor
        best_neighbor_index = np.argmin(neighbor_scores)
        best_neighbor_score = neighbor_scores[best_neighbor_index]

        # Accept the best neighbor if it improves the solution
        if best_neighbor_score < current_fitness:
            current_solution = neighbors[best_neighbor_index]
            current_fitness = best_neighbor_score

            # Track prediction flips
            new_prediction = predictions[best_neighbor_index]
            if current_prediction[0] != new_prediction[0]:
                nFlips[model] += 1
                if counter == 0:
                    flippedIterations[model].append(i+1)
                    counter += 1

    return current_solution, current_fitness

# Shared data structures for tracking flips
manager = Manager()
nFlips = manager.dict({1: 0, 2: 0})
flippedIterations = manager.dict({1: [], 2: []})

best_solutions_scores_3 = []
best_solutions_scores_4 = []

# Select a subset of test data
X_test_subset = X_test.values  # Convert DataFrame to NumPy for faster processing
y_test_subset = y_test.values

def task(X_test_sample, y_test_sample, model):
    """Runs stochastic hill climbing for a single sample (modifying only age & gender)."""
    initial_solution = X_test_sample.reshape(1, -1)
    final_solution, best_score = stochastic_hill_climber_specific(initial_solution, model, y_test_sample, num_iterations=10, num_neighbors=10)
    return X_test_sample, y_test_sample, final_solution, best_score

# Parallel Execution Using ThreadPoolExecutor
tasks = [(X_test_sample, y_test_sample, 1) for X_test_sample, y_test_sample in zip(X_test_subset, y_test_subset)]

with ThreadPoolExecutor() as executor:
    running_tasks = [executor.submit(task, *task_args) for task_args in tasks]
    for running_task in as_completed(running_tasks):
        best_solutions_scores_3.append(running_task.result())

# Repeat for Model 2
tasks = [(X_test_sample, y_test_sample, 2) for X_test_sample, y_test_sample in zip(X_test_subset, y_test_subset)]

with ThreadPoolExecutor() as executor:
    running_tasks = [executor.submit(task, *task_args) for task_args in tasks]
    for running_task in as_completed(running_tasks):
        best_solutions_scores_4.append(running_task.result())
```

```python
import pandas as pd
import numpy as np

def load_best_solutions(file_path):
    """
    Reads a CSV file, extracts solutions and checked values, and reconstructs
    the best_solutions_scores list.

    Parameters:
    - file_path (str): Path to the CSV file.

    Returns:
    - list: Reconstructed best_solutions_scores list with NumPy arrays.
    """
    df = pd.read_csv(file_path)

    # Ensure 'checked' column exists before dropping
    if 'checked' not in df.columns:
        raise KeyError(f"'checked' column not found in {file_path}")

    # Extract checked values (labels) as a NumPy array
    checked_values = df['checked'].to_numpy(dtype=int)

    # Extract feature values as a NumPy array
    solutions = df.drop(columns=['checked']).to_numpy(dtype=float)

    # Reshape solutions and construct tuples
    best_solutions_scores = [
        (None, y, sol.reshape(1, -1), None)  # Removed extra list wrapper []
        for y, sol in zip(checked_values, solutions)
    ]

    return best_solutions_scores

# Define file paths
file_paths = [
    "stochastic_hillclimber_model_1.csv",
    "stochastic_hillclimber_model_2.csv",
    "stochastic_hillclimber_model_3.csv",
    "stochastic_hillclimber_model_4.csv"
]

# Load solutions into respective lists
(best_solutions_scores_1,
 best_solutions_scores_2,
 best_solutions_scores_3,
 best_solutions_scores_4) = [load_best_solutions(fp) for fp in file_paths]
```

```python
from sklearn.ensemble import GradientBoostingClassifier
import numpy as np
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(15, 12))  # Increase height for better readability

for i, solutions in enumerate([best_solutions_scores_1, best_solutions_scores_2]):
    X_original, Y_labels, X_mutated, _ = zip(
        *[(x_sample, y_sample, solution[0], o_solution) for (x_sample, y_sample, solution, o_solution) in solutions])

    # Train Gradient Boosting Classifier
    gbc_original = GradientBoostingClassifier()
    gbc_original.fit(X_original, Y_labels)
    original_feature_importances = gbc_original.feature_importances_

    gbc_mutated = GradientBoostingClassifier()
    gbc_mutated.fit(X_mutated, Y_labels)
    mutated_feature_importances = gbc_mutated.feature_importances_

    # Get top 20 feature indices separately for original and mutated models
    top_indices_original = np.argsort(original_feature_importances)[-20:][::-1]
    top_indices_mutated = np.argsort(mutated_feature_importances)[-20:][::-1]

    # Extract corresponding feature names and values
    top_original_features = np.array(X_test.columns)[top_indices_original]
    top_mutated_features = np.array(X_test.columns)[top_indices_mutated]

    top_original_values = original_feature_importances[top_indices_original]
    top_mutated_values = mutated_feature_importances[top_indices_mutated]

    # Set y positions for both bars to be **one below the other**
    y_positions = np.arange(40)  # 40 slots: 20 for original + 20 for mutated

    # Plot bars (stacked vertically)
    axs[i].barh(y_positions[:20], top_original_values, height=0.4, label='Original', color='tab:blue')
    axs[i].barh(y_positions[20:], top_mutated_values, height=0.4, label='Mutated', color='tab:orange')

    # Set feature labels for each bar
    feature_labels = list(top_original_features) + list(top_mutated_features)
    axs[i].set_yticks(y_positions)
    axs[i].set_yticklabels(feature_labels, fontsize=10)  # Adjust font size for clarity

    axs[i].invert_yaxis()  # Highest importance at the top
    axs[i].set_xlabel('Feature Importance')
    axs[i].set_title(f'Model {i+1} (Top 20 Features)')
    axs[i].legend()

fig.suptitle('Top 20 Feature Importances: Original vs. Mutated', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
from tabulate import tabulate
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import numpy as np
import onnxruntime as rt

def calculate_metrics(y_true, y_pred):
    return accuracy_score(y_true, y_pred), precision_score(y_true, y_pred), recall_score(y_true, y_pred), f1_score(y_true, y_pred), roc_auc_score(y_true, y_pred)

# MODEL 1
new_session = rt.InferenceSession("Group19/model_1.onnx")

y_true = y_test.values
y_pred_original_1 = [get_predict_label_and_confidence(new_session, sample.reshape(1, -1))[0] for sample in X_test.values]
y_true_1_mutated = [y for _, y, _, _ in best_solutions_scores_1]
y_pred_1_mutated = [get_predict_label_and_confidence(new_session, solution)[0] for _, _, solution, _ in best_solutions_scores_1]

# MODEL 2
new_session = rt.InferenceSession("Group19/model_2.onnx")

y_pred_original_2 = [get_predict_label_and_confidence(new_session, sample.reshape(1, -1))[0] for sample in X_test.values]
y_true_2_mutated = [y for _, y, _, _ in best_solutions_scores_2]
y_pred_2_mutated = [get_predict_label_and_confidence(new_session, solution)[0] for _, _, solution, _ in best_solutions_scores_2]

# Calculate metrics
metrics_original_1 = calculate_metrics(y_true, y_pred_original_1)
metrics_mutated_1 = calculate_metrics(y_true_1_mutated, y_pred_1_mutated)
metrics_original_2 = calculate_metrics(y_true, y_pred_original_2)
metrics_mutated_2 = calculate_metrics(y_true_2_mutated, y_pred_2_mutated)

# Format metrics for model 1 and 2
metrics_str = "\nModel 1 Metrics:\n" + tabulate([
    ["Original", *metrics_original_1],
    ["Mutated", *metrics_mutated_1]
], headers=["Data", "Accuracy", "Precision", "Recall", "F1", "ROC"], tablefmt="grid")

metrics_str += "\n\nModel 2 Metrics:\n" + tabulate([
    ["Original", *metrics_original_2],
    ["Mutated", *metrics_mutated_2]
], headers=["Data", "Accuracy", "Precision", "Recall", "F1", "ROC"], tablefmt="grid")

# Compute confusion matrices
cm_original_1 = confusion_matrix(y_true, y_pred_original_1)
cm_mutated_1 = confusion_matrix(y_true_1_mutated, y_pred_1_mutated)
cm_original_2 = confusion_matrix(y_true, y_pred_original_2)
cm_mutated_2 = confusion_matrix(y_true_2_mutated, y_pred_2_mutated)

# Format confusion matrices as clear text (instead of forcing alignment)
confusion_matrix_str = "\n=== Confusion Matrices ===\n"

confusion_matrix_str += f"\nModel 1 Confusion Matrix (Original Data):\n{cm_original_1}\n"
confusion_matrix_str += f"\nModel 1 Confusion Matrix (Mutated Data):\n{cm_mutated_1}\n"

confusion_matrix_str += f"\nModel 2 Confusion Matrix (Original Data):\n{cm_original_2}\n"
confusion_matrix_str += f"\nModel 2 Confusion Matrix (Mutated Data):\n{cm_mutated_2}\n"

# Save results to a file
with open('stochastic_hillclimber_results1.txt', 'w') as f:
    f.write(metrics_str)
    f.write("\n\n")
    f.write(confusion_matrix_str)

print("Results saved successfully.")
```

```python
def get_age_neighbours(X_test_sample, y_test_sample, age_features):
    X_test_neighbours = []
    for feat in age_features:
        n1, n2, n3, n4, n5, n6, n7, n8 = [X_test_sample.copy() for _ in range(8)]
        og_feature_value = X_test_sample[feat].values[0]

        n1[feat] = max(0, min(100, og_feature_value + np.random.randint(-20, -5)))  
        n2[feat] = max(0, min(100, og_feature_value + np.random.randint(5, 20)))
        n3[feat] = max(0, min(100, og_feature_value + np.random.randint(-30, -10))) 
        n4[feat] = max(0, min(100, og_feature_value + np.random.randint(10, 30))) 
        n5[feat] = max(0, min(100, og_feature_value + np.random.randint(-50, -20))) 
        n6[feat] = max(0, min(100, og_feature_value + np.random.randint(20, 50)))
        n7[feat] = max(0, min(100, og_feature_value + np.random.randint(-70, -30))) 
        n8[feat] = max(0, min(100, og_feature_value + np.random.randint(30, 70))) 

        X_test_neighbours.extend([n1, n2, n3, n4, n5, n6, n7, n8])

    y_test_neighbours = [y_test_sample] * len(X_test_neighbours)
    
    return X_test_neighbours, y_test_neighbours

def get_ethnicity_neighbours(X_test_sample, y_test_sample, ethnicity_features):
    X_test_neighbours = []
    n_feat = len(ethnicity_features) 
    for feat in ethnicity_features:
        neigh = X_test_sample.copy()
        # For curr feature, flip the bool
        neigh[feat] = not neigh[feat].values[0]
        X_test_neighbours.append(neigh)
        
    y_test_neighbours = [y_test_sample] * len(X_test_neighbours)
    return X_test_neighbours, y_test_neighbours

def compute_fitness(conf_pairs_list, og_output_label):
    fitness = []
    rank1_key, rank2_key = None, None
    for i, conf_pair in enumerate(conf_pairs_list):
        if i == 0:
            rank1_key = max(conf_pair, key=conf_pair.get)
            rank2_key = min(conf_pair, key=conf_pair.get)
            val = round(conf_pair[rank1_key] - conf_pair[rank2_key], 3)
            fitness.append(val)
            continue
        neigh_rank1_key = max(conf_pair, key=conf_pair.get)
        neigh_rank2_key = min(conf_pair, key=conf_pair.get)
        if neigh_rank1_key != rank1_key:
            val = round(-conf_pair[neigh_rank1_key], 3)  
        else:
            val = round(conf_pair[neigh_rank1_key] - conf_pair[neigh_rank2_key], 3)
        fitness.append(val)
    return fitness

def get_neighbours_fitness(random_test_X, random_test_y,model):
    neighs_X, neighs_y = get_ethnicity_neighbours(random_test_X, random_test_y, ethnicity_features)
    new_test_X = random_test_X._append(neighs_X)             
    new_test_y = random_test_y._append(pd.Series(neighs_y))
    if model == "1":
        y_pred_onnx2 =  new_session1.run(None, {'X': new_test_X.values.astype(np.float32)})
    else: 
        y_pred_onnx2 =  new_session2.run(None, {'X': new_test_X.values.astype(np.float32)})
    model_confs_test = y_pred_onnx2[1]
    print(f"{len(model_confs_test)} confidence pairs on test data: {model_confs_test}")
    fitness_vals = compute_fitness(model_confs_test, random_test_y.values[0])
    print(f"Fitness values: {fitness_vals}")
    
    return new_test_X, new_test_y, fitness_vals
```

### Hill Climber for Ethnicity Features

```python
eps = 0.001    # TODO: possibly add eps
patience = 50  # If fitness value of neighbours is not WORSE than current solution for patience iterations, stop
count = 0
max_iter = 200

# TODO: Sample a random test sample from test set for which checked = 1
random_test_X = X_test.sample(n=1)
random_test_y = y_test[random_test_X.index]
print(type(random_test_X), random_test_X.shape, type(random_test_y), random_test_y.shape)
print(f"Random test sample has og label {random_test_y.values[0]}")

best_neigh_X = random_test_X    # Current best neighbour is set to initial random test sample
best_neigh_y = random_test_y
best_fitness_ethnic_vls = []   # 2d list of dict of 16 ethnicity values    
best_fitness_vals = []

# 4. Stopping criteria: none of the neighbours have a LOWER fitness value than the current solution (use eps)]
while True:
    count += 1
    max_iter -= 1
    print(f"\nIteration {count}")
    # 1. Start with random test sample: get model confidence for this sample
    new_test_X, new_test_y, fitness_vals = get_neighbours_fitness(best_neigh_X, best_neigh_y,"1")
    curr_fitness = fitness_vals[0]
    
    # Create dict with keys for all ethinicity features and corresponding values, for curr best sample
    curr_ethnicity_list = []
    for feat in ethnicity_features:
        curr_ethnicity_list.append(best_neigh_X[feat].values[0])
    best_fitness_ethnic_vls.append(curr_ethnicity_list)

    # Get indices of lowest fitness value in list
    min_fitness_idxs = [i for i, x in enumerate(fitness_vals) if x == min(fitness_vals)]
    # If none of the neighbours has a lower fitness value than the current solution, decrease patience
    if all(fitness_val >= curr_fitness for fitness_val in fitness_vals):
        print("None of the neighbours has lower fitness --- decreasing patience and picking random MIN")
        patience -= 1
        # We do pick a random sample out of all min values, to avoid getting stuck in local minima
        idx = np.random.choice(min_fitness_idxs[1:])     # TODO: idx 0 is excluded bc it gives issues
        best_neigh_X = new_test_X.iloc[idx].to_frame().transpose()
        best_neigh_y = new_test_y.iloc[idx]
        best_fitness_vals.append(fitness_vals[idx])
    elif len(min_fitness_idxs) == 1:                            # If there is only one other min value, update best neighbour                 
        best_neigh_X = new_test_X.iloc[min_fitness_idxs[0]].to_frame().transpose()
        best_neigh_y = new_test_y.iloc[min_fitness_idxs[0]]
        best_fitness_vals.append(fitness_vals[min_fitness_idxs[0]])
    else:                                                       # If there are multiple min values (incl. curr possibly), choose randomly
        idx = np.random.choice(min_fitness_idxs[1:])     # TODO: idx 0 is excluded bc it gives issues
        best_neigh_X = new_test_X.iloc[idx].to_frame().transpose()
        best_neigh_y = new_test_y.iloc[idx]
        best_fitness_vals.append(fitness_vals[idx])
    if patience == 0 or max_iter == 0:
        break

print(f"\n\n {best_fitness_ethnic_vls}")
print(f"Best fitness vals: {best_fitness_vals}")
```
